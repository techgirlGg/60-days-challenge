```python
import re
from collections import Counter
import matplotlib.pyplot as plt

# ---------------------------
# Sample Text (You Can Change)
# ---------------------------
text = """
This product is very good and this product is very useful.
I love this product. This is good and useful.
"""

# ---------------------------
# Clean Text
# ---------------------------
text = text.lower()
text = re.sub(r'[^a-z\s]', '', text)

# ---------------------------
# Word Frequency
# ---------------------------
words = text.split()
word_freq = Counter(words)

# ---------------------------
# Top 20 Words
# ---------------------------
top_20 = word_freq.most_common(20)

words_list = [w[0] for w in top_20]
freq_list = [w[1] for w in top_20]

print(top_20)

# ---------------------------
# Plot
# ---------------------------
plt.figure()
plt.bar(words_list, freq_list)
plt.xticks(rotation=45)
plt.title("Top 20 Words")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.show()

sentences = [
    "This product is very good",
    "I love this product",
    "This is useful and good",
    "Very good product quality"
]

text = " ".join(sentences)

with open("data.txt", "r", encoding="utf-8") as f:
    text = f.read()
```
