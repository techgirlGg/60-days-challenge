```python
import nltk
import re
import pandas as pd

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
quiz = [
    {
        "question": "What does NLP stand for?",
        "options": ["Natural Language Processing", "Neural Language Programming", "Natural Learning Process", "None"],
        "answer": "Natural Language Processing"
    },
    {
        "question": "Which library is commonly used for NLP in Python?",
        "options": ["NumPy", "NLTK", "Matplotlib", "TensorFlow"],
        "answer": "NLTK"
    },
    {
        "question": "What is tokenization?",
        "options": [
            "Removing stopwords",
            "Breaking text into words",
            "Converting text to lowercase",
            "Removing punctuation"
        ],
        "answer": "Breaking text into words"
    },
    {
        "question": "What is stemming?",
        "options": [
            "Root word extraction",
            "Sentence splitting",
            "POS tagging",
            "Translation"
        ],
        "answer": "Root word extraction"
    }
]

score = 0

for q in quiz:
    print("\n" + q["question"])
    for i, opt in enumerate(q["options"], 1):
        print(f"{i}. {opt}")
    
    user_ans = int(input("Enter option number: "))
    
    if q["options"][user_ans - 1] == q["answer"]:
        print("✅ Correct")
        score += 1
    else:
        print("❌ Wrong")

print(f"\nFinal Score: {score}/{len(quiz)}")
```

```python
text = """
NLP is amazing! It helps computers understand human language.
Text preprocessing is the first step in NLP pipeline.
"""
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    return text

cleaned_text = clean_text(text)
print("Cleaned Text:\n", cleaned_text)
```

```python
tokens = word_tokenize(cleaned_text)
print("Tokens:\n", tokens)
```

```python
stop_words = set(stopwords.words('english'))

filtered_tokens = [word for word in tokens if word not in stop_words]
print("After Stopword Removal:\n", filtered_tokens)
```

```python
stemmer = PorterStemmer()

stemmed_words = [stemmer.stem(word) for word in filtered_tokens]
print("Stemmed Words:\n", stemmed_words)
```

```python
lemmatizer = WordNetLemmatizer()

lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_tokens]
print("Lemmatized Words:\n", lemmatized_words)
```

```python
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    
    tokens = word_tokenize(text)
    
    stop_words = set(stopwords.words('english'))
    tokens = [w for w in tokens if w not in stop_words]
    
    stemmer = PorterStemmer()
    tokens = [stemmer.stem(w) for w in tokens]
    
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(w) for w in tokens]
    
    return tokens

print("\nFinal Output:")
print(preprocess_text(text))
```
