```python
documents = [
    "I love machine learning",
    "Machine learning is amazing",
    "I love coding"
]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Create BoW vectorizer
bow_vectorizer = CountVectorizer()

# Fit and transform the documents
bow_vectors = bow_vectorizer.fit_transform(documents)

# Convert to array
bow_array = bow_vectors.toarray()

# Print feature names
print("Feature Names:", bow_vectorizer.get_feature_names_out())

# Print BoW vectors
print("Bag of Words Vectors:\n", bow_array)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Create TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform
tfidf_vectors = tfidf_vectorizer.fit_transform(documents)

# Convert to array
tfidf_array = tfidf_vectors.toarray()

# Print feature names
print("Feature Names:", tfidf_vectorizer.get_feature_names_out())

# Print TF-IDF vectors
print("TF-IDF Vectors:\n", tfidf_array)
```
